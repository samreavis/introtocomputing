```python
#Starting to test the Amazon data
import pandas as pd
amazon = pd.read_csv('amazon.csv')
amazon
```

```python
amazon.shape
```

```python
amazon = amazon.fillna(0)
amazon.head()
```

```python
amazon["Total Charged"] = amazon["Total Charged"].str.replace('$','').astype(float)
amazon.head()
```

```python
amazon["Total Charged"].sum()
```

```python
amazon["Total Charged"].mean()
```

```python
amazon["Total Charged"].median()
#the median being lower than the mean shows there were a few expensive orders on this account
```

```python
amazon["Total Charged"].max()
```

```python
amazon["Total Charged"].min()
```

```python
amazon["Tax Charged"] = amazon["Tax Charged"].str.replace('$','').astype(float)
amazon.head()
```

```python
amazon["Tax Charged"].sum()
```

```python
amazon["Tax Charged"].sum() / amazon["Total Charged"].sum()
#The overall effective sales tax rate is about 5.29%
```

```python
#Analyzing Amazon Spending Over Time 
amazon['Order Date'] = pd.to_datetime(amazon['Order Date'])
amazon['Order Date'] = pd.to_datetime(amazon['Order Date']).dt.year
amazon
```

```python
amazon.plot.bar(x='Order Date', y='Total Charged', rot=90, figsize=(10,10))
```

```python
daily_orders = amazon.groupby('Order Date').sum()["Total Charged"]
daily_orders.plot.bar(figsize=(20,10),color='#61D199')
#2020 and 2021 appear to be the highest spending years, which should come as no surprise given the effects that COVID had on spending
```
